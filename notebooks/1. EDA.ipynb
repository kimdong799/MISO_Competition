```python
sales_rate_20 = pd.DataFrame(df_2021.groupby('서비스_업종_코드_명').mean().연령대_20_매출_비율)
sales_rate_20 = sales_rate_20.sort_values('연령대_20_매출_비율', ascending=False)

sales_rate_30 = pd.DataFrame(df_2021.groupby('서비스_업종_코드_명').mean().연령대_30_매출_비율)
sales_rate_30 = sales_rate_30.sort_values('연령대_30_매출_비율', ascending=False)
```

```python
topics = ['20대', '30대']
value_a = [12, 35]
value_b = [51, 28]

def create_x(t, w, n, d):
    return [t*x + w*n for x in range(d)]

value_a_x = create_x(2, 0.8, 1, 2)
value_b_x = create_x(2, 0.8, 2, 2)

ax = plt.subplot()
ax.bar(value_a_x, value_a, label='상향산업')
ax.bar(value_b_x, value_b, label='하향산업')
middle_x = [(a+b)/2 for (a,b) in zip(value_a_x, value_b_x)]
ax.set_xticks(middle_x)
ax.set_xticklabels(topics)
plt.legend()
plt.show()
```
