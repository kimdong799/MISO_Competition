```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

res = requests.get("http://corners.gmarket.co.kr/BestSellers?viewType=C&largeCategoryCode=100000035")
soup = BeautifulSoup(res.content, 'html.parser')

product_name = []
mydata = soup.find_all('a', "itemname")
for i in mydata:
    print(i.get_text())
    product_name.append(i.get_text())
    
product_price=[]
mydata = soup.find_all('div', "s-price")
for i in mydata:
    print(i.get_text())
    product_price.append(i.get_text())
    
product_dict = [i for i in zip(product_name, product_price)]

product_df = pd.DataFrame(product_dict)
product_df = product_df.drop([0,1,2,3,4])
product_df = product_df.reset_index()
product_df.drop('index', axis=1, inplace = True)
product_df.columns = ['상품명','가격']

product_df["가격"] = product_df["가격"].str.split("원").str[0]
product_df["가격"] = product_df["가격"].str.replace("[^0-9 ]","") 

product_df
```

```python
product_df.to_csv('../data/G마켓_유아의류_상품리스트.csv', encoding='utf-8-sig')
```
